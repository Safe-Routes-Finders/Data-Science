```python
import pandas as pd
```

```python
#Loading the data of accidents happening on Los Angeles
accidents = pd.read_csv("../data/Traffic_Collision_Data_from_2010_to_Present_Updated.csv")
accidents.shape
```

```python
#Converting date occured object column to datetime column
accidents['date_occurred'] = pd.to_datetime(accidents['Date Occurred'])
```

```python
#accidents at los angeles areas by year
accidents_year = pd.crosstab(accidents['Area Name'],accidents['date_occurred'].dt.year)
accidents_year.head()
```

```python
#adding the total accidents at a area for all the years
accidents_year['all_years'] = accidents_year.sum(axis=1)
```

```python
#Accidents on area by month of year
accidents_month = pd.crosstab(accidents['Area Name'],accidents['date_occurred'].dt.month)
accidents_month.head()
```

```python
#mering the number of accidents per year and number of accidents by month
accidents_agg = accidents_year.merge(accidents_month,left_index=True,right_index=True)
accidents_agg.head()
```

```python
accidents_agg = accidents_agg.reset_index()
```

```python
#Replacing column values for meaningful names
accidents_agg.columns = ['Area Name', 'yr_2010', 'yr_2011', 'yr_2012', 'yr_2013', 'yr_2014', 'yr_2015', 'yr_2016', 'yr_2017', 'yr_2018', 'yr_2019', 'all_years',
 'Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
accidents_agg.head()
```

```python
#Converting time to hour of time
accidents['Time Occurred'] = accidents['Time Occurred']//100
```

```python
#Accidents on areas by time of day
accidents_time = pd.crosstab(accidents['Area Name'],accidents['Time Occurred'])
accidents_time.head()
```

```python
accidents_agg = accidents_agg.set_index('Area Name')
accidents_agg.head()
```

```python
#merging with aggregate file
accidents_agg = accidents_agg.merge(accidents_time,left_index=True,right_index=True)
accidents_agg
```

```python
#dumping to csv
accidents_agg.to_csv('../data/accidents_aggregate_laareas.csv')
```
