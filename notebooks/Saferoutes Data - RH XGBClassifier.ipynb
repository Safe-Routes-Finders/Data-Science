```python
#NYC Bounding Boxes
UL_LAT = 40.916178
UL_LON = -74.25909
LR_LAT = 40.477399
LR_LON = -73.700181
```

```python
df[((df['LATITUDE'] > LR_LAT))&((df['LATITUDE'] < UL_LAT)) & ((df['LONGITUD'] > UL_LON)) & ((df['LONGITUD'] < LR_LON)) ]
```

```python
df = df[((df['LATITUDE'] > LR_LAT))&((df['LATITUDE'] < UL_LAT)) & ((df['LONGITUD'] > UL_LON)) & ((df['LONGITUD'] < LR_LON)) ]
```

```python
df['LATITUDE'] = round(df['LATITUDE'],4)
```

```python
df['LONGITUD'] = round(df['LONGITUD'],4)
```

```python
df['LONGITUD'] = df['LONGITUD'].astype(str)
```

```python
df['LATITUDE'] = df['LATITUDE'].astype(str)
```

```python
df['LATLON'] = df[['LATITUDE', 'LONGITUD']].apply(lambda x: ''.join(x), axis=1)
```
