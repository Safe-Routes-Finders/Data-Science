```python
import pandas as pd
```

```python
#Loading the data of accidents happening on Los Angeles
accidents = pd.read_csv("../data/Traffic_Collision_Data_from_2010_to_Present_Updated.csv")
accidents.shape
```

```python
accidents.tail()
```

```python
#Finding the unique number of Intersections with number of accidents
accidents['Location'].value_counts()
```

```python
#Converting date occured object column to datetime column
accidents['date_occurred'] = pd.to_datetime(accidents['Date Occurred'])
```

```python
accidents['date_occurred'].tail()
```

```python
#Minimum and Maximum dates in the data
accidents['date_occurred'].min(),accidents['date_occurred'].max()
```

```python
#checking for number of unique accident locations
accidents['Location'].nunique()
```

```python
#accidents at location by year
accidents_year = pd.crosstab(accidents['Location'],accidents['date_occurred'].dt.year)
accidents_year.head()
```

```python
#adding the total accidents at a location for all the years
accidents_year['all_years'] = accidents_year.sum(axis=1)
```

```python
accidents_year.tail()
```

```python
#Accidents on location by month of year
accidents_month = pd.crosstab(accidents['Location'],accidents['date_occurred'].dt.month)
accidents_month.head()
```

```python
#mering the number of accidents per year and number of accidents by month
accidents_agg = accidents_year.merge(accidents_month,left_index=True,right_index=True)
accidents_agg.head()
```

```python
accidents_agg = accidents_agg.reset_index()
```

```python
#Replacing column values for meaningful names
accidents_agg.columns = ['Location', 'yr_2010', 'yr_2011', 'yr_2012', 'yr_2013', 'yr_2014', 'yr_2015', 'yr_2016', 'yr_2017', 'yr_2018', 'yr_2019', 'all_years',
 'Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
accidents_agg.head()
```

```python
#Creating latitude and longitude columns for the accident locations
locations = accidents_agg['Location']
latlonglist = locations.str.replace('(','').str.replace(')','').to_list()
latitude = []
longitude = []
for lat in latlonglist:
    combine = lat.split(',')
    latitude.append(combine[0])
    longitude.append(combine[1])
accidents_agg['latitude']=latitude
accidents_agg['longitude']=longitude
```

```python
#checking for columns before dumping the dataframe to csv
accidents_agg.head()
```

```python
#dumping to csv
accidents_agg.to_csv('../data/accidents_aggregate_la.csv',index=False)
```
